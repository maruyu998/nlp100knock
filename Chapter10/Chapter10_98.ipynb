```python
def spacy_train(params, devicestr:str, dicsource:str, trainsource:str, tag:str="", examples=["私以外私じゃないの", "私は猫です。", "日本の水墨画を一変させた。"], model=None):
    import pickle, spacy
    from string import ascii_letters
    from random import choices
    with open(f'model_logs/{dicsource}_spacy_ja.dic', 'rb') as f: ja_dic = pickle.load(f)
    with open(f'model_logs/{dicsource}_spacy_en.dic', 'rb') as f: en_dic = pickle.load(f)
    ja_dicsize, en_dicsize = max(ja_dic.values())+1, max(en_dic.values())+1
    model, optimizer, scheduler, device = get_elms(params, devicestr, ja_dicsize, en_dicsize, model)
    translator = Translator(ja_dic, en_dic, spacy.load('ja_ginza'), spacy.load('en_core_web_sm'))
    trainset = DataSet(trainsource, maxlength=60, pathkey=f"{dicsource}_spacy_ids")
    devset = DataSet("dev", maxlength=60, pathkey=f"{dicsource}_spacy_ids")
    bleu_cb = BleuCallback(lambda seq: " ".join([translator.en_dic_rev[s] if s!=12 else ''.join(choices(ascii_letters,k=5)) for s in seq]))
    es_cb = EarlyStopping(count=10)
    save_cb = Save(model,f'model_logs/98_{dicsource}_spacy_{trainsource}/')
    tb_cb = Tensorboard('./tb_logs/', f'98_{dicsource}_spacy_{trainsource}', bleu_cb)
    print_cb = Print(model, translator, device, examples=examples, early_stopping=es_cb, bleu_callback=bleu_cb)
    callbacks = [es_cb,bleu_cb,save_cb,tb_cb,print_cb]
    trainer = Trainer(translator, trainset, devset, device=device)
    return model, lambda:trainer.run(model, optimizer, epoch=300, device=device, steps=10000, batch_size=params['batch_size'], shuffle=True, scheduler=scheduler, callbacks=callbacks)
```

```python
batch_size = 64*2
lrate_coef = 25000/batch_size*2
params = get_param(lrate_coef=lrate_coef)
params['batch_size'] = batch_size
trainsource = "mix"
dicsource = "kyoto-train"
model, runner = spacy_train(params, 'cuda:1', dicsource, trainsource)
```

```python
def subword_train(vocab_size, params, devicestr:str, trainsource:str, spmsource:str, tag:str="",
                  examples=["私以外私じゃないの", "私は猫です。", "日本の水墨画を一変させた。"], 
                  tgtfiles=["composed","mix","kyoto-train","kyoto-dev","kyoto-test"]):
    sp_tokenize(vocab_size, source=spmsource, tgtfiles=tgtfiles)
    sp_idation(vocab_size, source=spmsource)
    ja_sp, en_sp = load_spm(vocab_size, source=spmsource)
    ja_dic, en_dic = sp_load_dict(vocab_size, source=spmsource)
    ja_dicsize, en_dicsize = max(ja_dic.values())+1, max(en_dic.values())+1
    model, optimizer, scheduler, device = get_elms(params, devicestr, ja_dicsize, en_dicsize)
    translator = Translator(ja_dic, en_dic, ja_sp.EncodeAsPieces, en_sp.EncodeAsPieces)
    trainset = DataSet(trainsource, maxlength=65, pathkey=f"{spmsource}_{vocab_size}_ids")
    devset = DataSet("dev", maxlength=65, pathkey=f"{spmsource}_{vocab_size}_ids")
    es_cb = EarlyStopping(count=10)
    print_cb = Print(model, translator, device, examples=examples, earlyStopping=es_cb)
    save_cb = Save(model,f'model_logs/98_{vocab_size}_{tag}/')
    tb_cb = Tensorboard('./tb_logs/', f'98_{vocab_size}_{tag}')
    trainer = Trainer(translator, trainset, devset, device=device)
    return lambda:trainer.run(model, optimizer, epoch=300, device=device, batch_size=params['batch_size'], shuffle=True, scheduler=scheduler, callbacks=[es_cb,print_cb,save_cb,tb_cb], eval_double_bleu=False)
```
